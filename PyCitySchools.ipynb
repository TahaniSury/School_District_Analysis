```python
# Install numpy using conda install numpy or pip install numpy. 
# Step 1. Import numpy as np.
import numpy as np
```

```python
#  Use the loc method on the student_data_df to select all the reading scores from the 9th grade at Thomas High School and replace them with NaN.

student_data_df.loc[(student_data_df["school_name"] == "Thomas High School") & (student_data_df["grade"] == "9th") & (student_data_df["reading_score"] > 0), ["reading_score"]] = np.nan
#  Refactor the code in Step 2 to replace the math scores with NaN.
student_data_df.loc[(student_data_df["school_name"] == "Thomas High School") & (student_data_df["grade"] == "9th") & (student_data_df["math_score"] > 0), ["math_score"]] = np.nan

# (count for Thomas High School and 9th graders is 461 except for math and reading scores which is 'Nan')
# student_data_df.loc[(student_data_df["school_name"] == "Thomas High School") & (student_data_df["grade"] == "9th")].count()
student_data_df.tail(10)
```

```python
# Combine the data into a single dataset.
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
# school_data_complete_df.head()
# district_summary_df: Get the total number of students.
student_count = school_data_complete_df["Student ID"].count()
# student_count
# district_summary_df: Calculate the total number of schools.
school_count = school_data_df["school_name"].count()
# school_count
# district_summary_df: Calculate the total budget.
total_budget = school_data_df["budget"].sum()
# total_budget
# Replacing all 'Nan' values in math_score and reading_score column to 0 for further calculations, since with Nan, it would not consider number of students with Nan value in columns
final_school_data_complete_df = school_data_complete_df
final_school_data_complete_df['math_score'] = final_school_data_complete_df['math_score'].fillna(0)
final_school_data_complete_df['reading_score' ]= final_school_data_complete_df['reading_score'].fillna(0)
# final_school_data_complete_df.tail(10)
```

```python
# district_summary_df: Calculate the average reading score.
average_reading_score = school_data_complete_df["reading_score"].mean()
# average_reading_score
# district_summary_df: Calculate the average math score.
average_math_score = final_school_data_complete_df["math_score"].mean()
# average_math_score
# district_summary_df: Get all the students who are passing math in a new DataFrame to get % students passing in math.
passing_math = final_school_data_complete_df[final_school_data_complete_df["math_score"] >= 70]
# passing_math.head()
# district_summary_df: Get all the students that are passing reading in a new DataFrame to get % students passing in reading.
passing_reading = final_school_data_complete_df[final_school_data_complete_df["reading_score"] >= 70]
# passing_reading.head()
# district_summary_df: Getting count of students passing math to get % students passing in math.
passing_math_count = passing_math["student_name"].count()
# passing_math_count
# district_summary_df: Getting count of students passing math to get % students passing in reading.
passing_reading_count = passing_reading["student_name"].count()
# passing_reading_count
```

```python
# district_summary_df: Calculate the percent that passed math.
passing_math_percentage = passing_math_count / float(student_count) * 100
# passing_math_percentage
# district_summary_df: Calculate the percent that passed reading.
passing_reading_percentage = passing_reading_count / float(student_count) * 100
# passing_reading_percentage
# district_summary_df:Calculate the students who passed both math and reading to get % students passing overall. 
passing_math_reading = final_school_data_complete_df[(final_school_data_complete_df["math_score"] >= 70) & (final_school_data_complete_df["reading_score"] >= 70)]
# passing_math_reading.head()
# district_summary_df :Calculate the number of students who passed both math and reading  to get % students passing overall.
overall_passing_math_reading_count = passing_math_reading["student_name"].count()
# overall_passing_math_reading_count
# district_summary_df :Calculate the overall passing percentage n both math and reading.
overall_passing_percentage = overall_passing_math_reading_count / student_count * 100
# overall_passing_percentage
# district_summary_df: Adding a list of values with keys to create a new DataFrame.
district_summary_df = pd.DataFrame([{"Total Schools": school_count,
                                    "Total Students": student_count, 
                                    "Total Budget": total_budget, 
                                    "Average Math Score": average_math_score, 
                                    "Average Reading Score": average_reading_score, 
                                    "% Passing Math": passing_math_percentage, 
                                    "% Passing Reading": passing_reading_percentage, 
                                    "% Overall Passing": overall_passing_percentage}])
# district_summary_df
```

```python
# district_summary_df: Formatting the columns.
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)
district_summary_df
```

```python
# PER_SCHOOL_SUMMARY_DF
# _school_summary_df: Determine the school type.
per_school_types = school_data_df.set_index(["school_name"])["type"]
# per_school_types
# per_school_summary_df: Calculate the total student count.
per_school_counts = final_school_data_complete_df["school_name"].value_counts()
# per_school_counts
# per_school_summary_df: Calculate the total school budget.
per_school_budget = school_data_df.set_index(["school_name"])["budget"]
# per_school_budget
# per_school_summary_df: Calculate the per capita spending.
per_school_capita = per_school_budget / per_school_counts
# per_school_capita
# per_school_summary_df: Calculate the average test scores of math.
per_school_math = final_school_data_complete_df.groupby(["school_name"]).mean()["math_score"]
# per_school_math
# per_school_summary_df: Calculate the average test scores of reading.
per_school_reading = final_school_data_complete_df.groupby(["school_name"]).mean()["reading_score"]
# per_school_reading
# per_school_summary_df: Calculate the passing scores by creating a filtered DataFrame to get % students passing math.
per_school_passing_math = final_school_data_complete_df[(school_data_complete_df["math_score"] >= 70)]
# per_school_passing_math.tail(10)
# per_school_summary_df: Calculate the passing scores by creating a filtered DataFrame to get % students passing reading.
per_school_passing_reading = final_school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)]
# per_school_passing_reading.tail(10)
```

```python
#  per_school_summary_df: Calculate the number of students passing math by school to get % students passing math.
per_school_passing_math = per_school_passing_math.groupby(["school_name"]).count()["student_name"]
# per_school_passing_math
# Step B8(iv) of per_school_summary_df: Calculate the number of students passing reading by schoolto get % students passing reading.
per_school_passing_reading = per_school_passing_reading.groupby(["school_name"]).count()["student_name"]
# per_school_passing_reading
# Step B8(v) of per_school_summary_df: Calculate the percentage of students passing math per school.
per_school_passing_math = per_school_passing_math / per_school_counts * 100
# per_school_passing_math
# Step B8(vi) of per_school_summary_df: Calculate the percentage of students passing reading per school.
per_school_passing_reading = per_school_passing_reading / per_school_counts * 100
# per_school_passing_reading
# Step B9(i) of per_school_summary_df: Calculate the students who passed both math and reading to get % students passing overall.
per_passing_math_reading = final_school_data_complete_df[(final_school_data_complete_df["math_score"] >= 70) & (final_school_data_complete_df["reading_score"] >= 70)]
# per_passing_math_reading.tail(10)
# Step B9(ii) of per_school_summary_df: Calculate the number of students who passed both math and reading to get % students passing overall.
per_passing_math_reading = per_passing_math_reading.groupby(["school_name"]).count()["student_name"]
# per_passing_math_reading
# Step B9(iii) of per_school_summary_df: Calculate the overall passing percentage.
per_overall_passing_percentage = per_passing_math_reading / per_school_counts * 100
# per_overall_passing_percentage
```

```python
# per_school_summary_df: Adding a list of values with keys to create a new DataFrame.
per_school_summary_df = pd.DataFrame({
             "School Type": per_school_types,
             "Total Students": per_school_counts,
             "Total School Budget": per_school_budget,
             "Per Student Budget": per_school_capita,
             "Average Math Score": per_school_math,
           "Average Reading Score": per_school_reading,
           "% Passing Math": per_school_passing_math,
           "% Passing Reading": per_school_passing_reading,
           "% Overall Passing": per_overall_passing_percentage})
# per_school_summary_df
```
